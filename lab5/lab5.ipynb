### Numerical Integration

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_1 = lambda x: 1/np.power(x, 2)
f_2 = lambda x: 1 / np.log(x)
```

```python
def plot_f(xs, fs, label, points=None, area=None):
    fig1, ax1 = plt.subplots(figsize=(10,10))
    ax1.plot(xs, fs, label=label)
    if points is not None:
        ax1.scatter(points[0], points[1], color="r", s=50)
    ax1.grid(True, which='both')
    ax1.axhline(y=0, color='k')
    ax1.axvline(x=0, color='k')
    if area is not None:
        ax1.fill_between(*area)
    plt.legend()
    
def plot_func(func, x_min, x_max, N, label, points=None, area=None):
    xs = np.linspace(x_min, x_max, N)
    fs = func(xs)
    if area is None:
        draw_area = None
    if area is not None:
        draw_area = area
    if area == "default":
        draw_area = (xs, 0, fs)
    plot_f(xs, fs, label, points, draw_area)
    return xs, fs
```

```python
x1, f1 = plot_func(f_1, -2, 2, 100, label='$ \dfrac{1}{x^2} $')
```

```python
plot_f(x1, f1, label='$ \dfrac{1}{x^2} $', area=(x1,0,f1))
```

```python
x2, f2 = plot_func(f_2, 0, 2, 100, '$ \dfrac{1}{\ln{x}} $')
```

```python
plot_f(x2, f2, label='$ \dfrac{1}{\ln{x}} $', area=(x2[x2 >= 1], 0, f2[x2 >= 1]))
```

```python
f_3 = lambda x: 1/x
```

```python
x3, f3 = plot_func(f_3, 1, 2, 100, '$ \dfrac{1}{\ln{x}} $')
```

```python
f_3_d2 = lambda x: 2/np.power(x,3)
```

```python
x3d, f3d = plot_func(f_3_d2, 1, 2, 100, '$ \dfrac{1}{\ln{x}} $')
```
